```python
import requests as r
import pandas as pd

GENERAL = 'https://fantasy.premierleague.com/api/bootstrap-static/'



request = r.get(GENERAL)
all_data = request.json()
```

```python
player_columns = ['can_transact', 'can_select', 'chance_of_playing_next_round',
       'chance_of_playing_this_round', 'dreamteam_count', 'element_type',
       'first_name', 'form', 'id', 'in_dreamteam', 'news', 'news_added',
       'now_cost', 'points_per_game', 'second_name', 'selected_by_percent',
       'squad_number', 'status', 'team', 'total_points', 'transfers_in',
       'transfers_in_event', 'transfers_out', 'transfers_out_event',
       'value_form', 'value_season', 'web_name', 'opta_code', 'minutes',
       'goals_scored', 'assists', 'clean_sheets', 'goals_conceded',
       'own_goals', 'penalties_saved', 'penalties_missed', 'yellow_cards',
       'red_cards', 'saves', 'bonus', 'bps', 'influence', 'creativity',
       'threat', 'ict_index', 'clearances_blocks_interceptions', 'recoveries',
       'tackles', 'defensive_contribution', 'starts', 'expected_goals',
       'expected_assists', 'expected_goal_involvements',
       'expected_goals_conceded', 'corners_and_indirect_freekicks_order',
       'direct_freekicks_order', 'penalties_order', 'expected_goals_per_90',
       'saves_per_90', 'expected_assists_per_90',
       'expected_goal_involvements_per_90', 'expected_goals_conceded_per_90',
       'goals_conceded_per_90', 'starts_per_90', 'clean_sheets_per_90',
       'defensive_contribution_per_90']

team_columns = ['code', 'id', 'name', 'played', 'points',
       'position', 'short_name', 'strength', 'strength_overall_home', 'strength_overall_away',
       'strength_attack_home', 'strength_attack_away', 'strength_defence_home',
       'strength_defence_away']

position_columns = ['id', 'singular_name',
       'singular_name_short', 'squad_select', 'squad_min_play', 'squad_max_play',
       'element_count']
```

```python

players = pd.DataFrame(all_data['elements'])
players = players[player_columns]
players.to_csv('raw_data/players.csv', index=False)

teams = pd.DataFrame(all_data['teams'])
teams = teams[team_columns]
teams.to_csv('raw_data/teams.csv', index=False)

positions = pd.DataFrame(all_data['element_types'])
positions = positions[position_columns]
positions.to_csv('raw_data/positions.csv', index=False)

print("Players: ", players.shape)
print("Teams: ", teams.shape)
print("Positions: ", positions.shape)
```
